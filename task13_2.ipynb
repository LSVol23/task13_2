```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

def find_keywords(file_path, top_n=5):
    df = find_keywords("cafe_reviews.csv")

    if 'normalized_text' not in df.columns:
        raise ValueError("В файле отсутствует столбец 'normalized_text'")

    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(df['normalized_text'])

    feature_names = vectorizer.get_feature_names_out()

    def extract_keywords(row_vector):
        scores = row_vector.toarray().flatten()
        indices = scores.argsort()[::-1][:top_n]
        return [(feature_names[i], float(scores[i])) for i in indices if scores[i] > 0]

    df['key_words'] = [extract_keywords(tfidf_matrix[i]) for i in range(tfidf_matrix.shape[0])]

    return df
```
